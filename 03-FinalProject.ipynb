## Abstract

This project investigates the factors that influence pricing in U.S. athleticwear by analyzing product attributes of Adidas products. Using two merged datasets containing 3,454 observations of Adidas athletic apparel and 9,300 observations of footwear, we examined how gender-target, product type, color, country of origin, and popularity (measured by review counts) affect both listing and sale prices. 

We chose sales price the price type we we wanted to analyze after performing exploratory data analysis and inferential statistics (ANOVA) to see how different factors impacted pricing, and observed that listing price seemed to be a baseline for sales price, from which other product characteristics factored into determining the final sales price. 

We used hedonic pricing methodology, an economic technique that breaks down a good into its constituent characteristics to determine it's value, in our analysis.  This allowed us to perform multiple regression analysis, where we found that product type and gender-target were the most significant predictors of athleticwear pricing. 

This was contrary to our initial hypothesis that material and origin would be primary drivers of price, as our results showed that footwear commanded significantly higher prices than clothing or accessories, and products marketed to women showed different pricing patterns compared to men's products. Modest gender-based pricing disparities were evident, with women's products showing higher listing prices but similar sale prices across categories. Product type, as well as country were influential factors in determining pricing as well, which we discovered in our hedonic analysis.  These findings have important implications for understanding pricing strategies in the athleticwear industry, and raise questions about potential gender-based pricing practices and consumer behaviors. 

It is important to note that our analysis is limited to Adidas products in the U.S. market and cannot be generalized to the broader athleticwear industry, and that it does not reveal the entire truth of athleticwear pricing, even just for Adidas.  Our data was limited both in size (not enough data for a robust analysis) and in scope (missing information that would be significant to determining price, according to background research), and so we were not able to conduct an deep enough study to confidently yield broadly applicable results.

## Background and Prior Work

The prevalence of athletic wear, whether worn for leisure or actual activity, has been on the rise.  In fact, the athletic wear market now represents a substantial and rapidly growing segment of the global apparel industry, with the North American market alone valued at over 26 billion dollars in 2024, and a projected value of 677.26 billion dollars by 2030.  Among other things, this rapid growth can be attributed to a culture that is increasingly health-conscious, and the not-unrelated merging of fashion and functionality, dubbed "athlesiure". As casually-worn athletic wear becomes increasingly mainstream, however, there is the question of cost - specifically, what factors make it cost what is does? 

  
Now, the question of what drives apparel pricing has been examined before, at great depth, by economists around the world.  Multiple countries, including the United States, have used the hedonic model to calculate the Consumer Price Index, which is then used to track inflation for apparel in their countries. The hedonic model uses regression to "estimate a value for each characteristic of the item", which then all sum to the total price of the item. <a name="https://www.bls.gov/cpi/quality-adjustment/questions-and-answers.htm"></a>[<sup>1</sup>](#https://www.bls.gov/cpi/quality-adjustment/questions-and-answers.htm) The use of this technique allows researchers to make adjustments to account for seasonal changes and evolving product characteristics, essentially isolating the contribution of individual attributes to overall product pricing. 


One factor, material composition, represents one of the most fundamental drivers of garment pricing, with significant research demonstrating its importance. In the active wear market, synthetic materials were very popular, supported by widespread use of polyester, which has moisture-wicking properties and is durable, important qualities for active wear to have.<a name="https://www.marketresearchfuture.com/reports/sportswear-fabric-market-37412"></a>[<sup>2</sup>](#https://www.marketresearchfuture.com/reports/sportswear-fabric-market-37412)However, consumer preferences are evolving with the times, and natural materials like organic cotton and bamboo blends are becoming more popular as eco-friendly and sustainable products, since sustainability has become a more mainstream concern.  The price consumers are willing to pay for sustainable materials has been documented before. Research on outdoor apparel using hedonic pricing models found that material type had a statistically significant impact on item price, with different materials like cotton, polyester, wool, and nylon valued at different price points based on their breathability, durability, and environmental impact.<a name="https://www.researchgate.net/publication/332530663_The_Outdoor_Apparel_IndustryMeasuring_the_Premium_for_Sustainability_with_a_Hedonic_Pricing_Model"></a>[<sup>3</sup>](#https://www.researchgate.net/publication/332530663_The_Outdoor_Apparel_IndustryMeasuring_the_Premium_for_Sustainability_with_a_Hedonic_Pricing_Model) Not to mention, raw material price fluctuations also significantly impact manufacturing costs for brands, with materials such as organic cotton subject to changes in supply and demand as well as energy and transportation costs. This cost pressure directly influences retail pricing strategies, as manufacturers must balance quality expectations with competitive pricing. 

Another important consideration for pricing that transcends the realm of clothing is the "Pink Tax", or gender based pricing.  This is a recorded phenomenon, even reported on by the U.S. Congress's Joint Economic Committee, which notes that "... the frequency with which female consumers find themselves paying higher prices for gender-specific goods and services effectively becomes a tax on being a woman".<a name="https://www.jec.senate.gov/public/_cache/files/8a42df04-8b6d-4949-b20b-6f40a326db9e/the-pink-tax---how-gender-based-pricing-hurts-women-s-buying-power.pdf"></a>[<sup>4</sup>](#https://www.jec.senate.gov/public/_cache/files/8a42df04-8b6d-4949-b20b-6f40a326db9e/the-pink-tax---how-gender-based-pricing-hurts-women-s-buying-power.pdf) Gender-targeted marketing and its relationship to pricing represents one of the most controversial aspects of apparel pricing research, with another 2015 study by the New York City Department of Consumer Affairs finding that women's clothing was 8% more expensive than men's clothing on average.<a name="https://www.nyc.gov/assets/dca/downloads/pdf/partners/Study-of-Gender-Pricing-in-NYC.pdf"></a>[<sup>5</sup>](#https://www.nyc.gov/assets/dca/downloads/pdf/partners/Study-of-Gender-Pricing-in-NYC.pdf) Not only that, but women's clothing imports face higher average tariff rates than men's clothing, with women's clothing at 15.1% compared to 11.9% for men.<a name="http://bush.tamu.edu/wp-content/uploads/2020/07/V6-3-Tariff-Discrimination-Takeaway.pdf"></a>[<sup>6</sup>](#http://bush.tamu.edu/wp-content/uploads/2020/07/V6-3-Tariff-Discrimination-Takeaway.pdf) However, other research challenges the notion of gender-based pricing. Studies like these find that most comparable products differed in their ingredient compositions, suggesting that price differences may reflect genuine product differentiation rather than discriminatory pricing. <a name="https://pubsonline.informs.org/doi/pdf/10.1287/mksc.2023.1452"></a>[<sup>7</sup>](#https://pubsonline.informs.org/doi/pdf/10.1287/mksc.2023.1452) The debate centers on distinguishing between legitimate reasons in pricing difference and discriminatory practices, something we wish to also explore on a smaller level, in our analysis of what determines the cost of athletic wear. 


Country of origin (COO) represents another well-established driver of consumer perceptions and pricing in apparel markets. The COO effect describes how consumers' attitudes, perceptions, and purchasing decisions are influenced by products' country of origin labeling, which may refer to where a brand is based, where products are designed, or where they are manufactured.  Research has empirically demonstrated that the COO effect has significant price-related consequences, with brands possessing favorable COO associations able to charge price premiums over and above those attributed to observable product differentiation <a name="https://link.springer.com/article/10.1007/s11747-012-0318-1"></a>[<sup>8</sup>](#https://link.springer.com/article/10.1007/s11747-012-0318-1). The strength of COO effects varies considerably by product category and consumer characteristics. In apparel specifically, research examining clothing from multiple countries found that COO was related to assessment of product quality, but when evaluating purchase likelihood, COO seemed not to be as important, suggesting a more complex situation.<a name="https://www.researchgate.net/publication/318283638_The_Country-of-Origin_Construct_and_Its_Effect_on_Consumer_Behavior_A_Review_of_Selected_Literature_and_Proposed_Future_Research_Directions"></a>[<sup>9</sup>](#https://www.researchgate.net/publication/318283638_The_Country-of-Origin_Construct_and_Its_Effect_on_Consumer_Behavior_A_Review_of_Selected_Literature_and_Proposed_Future_Research_Directions)

The influence of popularity metrics on pricing is a newer area of research enabled by e-commerce platforms, which we choose to measure in the form of customer reviews and ratings in the absence of actual total purchasing data. Consumer surveys consistently shows that ratings and reviews are considered very important, with up to 98% of shoppers saying reviews are an "essential resource when making purchase decisions".<a name="https://www.powerreviews.com/power-of-reviews-2023/"></a>[<sup>10</sup>](#https://www.powerreviews.com/power-of-reviews-2023/) Research also demonstrates that products with five reviews have 270% greater purchase likelihood than products with no reviews, but these benefits diminish rapidly after the first five reviews, according to Northwestern.<a name="https://spiegel.medill.northwestern.edu/how-online-reviews-influence-sales/"></a>[<sup>11</sup>](#https://spiegel.medill.northwestern.edu/how-online-reviews-influence-sales/) Conjoint analysis research also indicates that "consumers considered star rating, number of reviews, online shopping platform, and premium reviews to be more important than price when making an online shopping decision". <a name="https://www.nature.com/articles/s41599-023-02277-7"></a>[<sup>12</sup>](#https://www.nature.com/articles/s41599-023-02277-7) However, there are many factors that differentiate from study to study, and so we would like to uncover ourselves how perceived popularity influence product pricing in the realm of active wear.

### PRELIMINARY ANALYSIS : INFERENTIAL STATISTICS - LOOKING AT GENDER TARGET AND PRICING

We chose gender target as our focus for looking at pricing due to our discoveries in EDA.  Namely, it was interesting that gender target seemed to be a factor that influenced pricing across sales and listing price, as well as when we factored in product type and country.  We looked at both sales and listing price to ensure that our inital idea of sales price being impacted by gender was a plausible one, as well as to see if listing price had any statistically significant relationship to gender target.  

From EDA, we already knew that sales price was lower than listing price, and so this section helped dictate whether our initial observation of gender being a factor in pricing was a valid one or not. This analysis also helped solidify our decison to predict sales price, using listing price as a baseline, and other characteristics of the products to contribute to the final sales price, as it showed there was a statistically significant difference in sales price but not listing price when it came to gender, indicating to us that there might be other factors that contribute to overall sales price as well. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

### loading our dataset
merged = pd.read_csv('data/02-processed/merged.csv')
df_model = merged.copy()
df_gender = df_model.copy()

print("\n\nPRELIMINARY ANALYSIS - INFERENTIAL STATISTICS : LOOKING AT GENDER TARGET AND PRICING")
print("-"*80)


### Calculate summary statistics for gender categories so that we can see the averages, etc. 
print("\n ------ Summary Statistics ------")
summary_stats = df_gender.groupby('Gender')[['Listing Price', 'Sale Price']].agg(['mean', 'std'])
print(summary_stats)

### Separating data into Men's vs Women's vs Unisex
mens = df_gender[df_gender['Gender'] == "Men's product"]
womens = df_gender[df_gender['Gender'] == "Women's product"]
unisex = df_gender[df_gender['Gender'] == "Unisex product"]


### Running an ANOVA test for gender groups, to check for any significant differences in 
### pricing (sales and listing)
print("\n\n--- ANOVA Test (Across Men, Women, and Unisex) ---\n")
f_stat, p_val = stats.f_oneway(
    mens['Listing Price'],
    womens['Listing Price'],
    unisex['Listing Price']
)
print(f"Listing Price: F-statistic = {f_stat}, p-value = {p_val}")
if p_val < 0.05:
    print(f"     At least one group is significantly different (p < 0.05)\n")
else:
    print(f"     No significant differences between groups (p >= 0.05)\n")

f_stat, p_val = stats.f_oneway(
    mens['Sale Price'],
    womens['Sale Price'],
    unisex['Sale Price']
)
print(f"Sale Price: F-statistic = {f_stat}, p-value = {p_val}")
if p_val < 0.05:
    print(f"     At least one group is significantly different (p < 0.05)\n")
else:
    print(f"     No significant differences between groups (p >= 0.05)]n")


print("\nIt looks like gender isn't a significant factor in listing price, but it may be in sales price.  \n\nWe will investigate this further by using sales price as our focus for prediction, rather than \nlisting price.  \n\nListing price may be a baseline for sales price, as noted above in the EDA section, so we will \nuse it as a variable in our prediction model.\n\n")
```

### ANAYLSIS 1: HEDONIC ANALYSIS - LOOKING AT WHAT FACTORS IMPACT SALE PRICE

Hedonic pricing is an economic method used to estimate the value of a good by analyzing the relationship between its price and its various characteristics. By using statistical analysis on product data, we can determine the implicit value of features.  Since we are utilizing hedonic pricing for this analysis, regression is the ideal method of choice.

Here, we try to simulate this using ordinary least squares regression, done manually so that we can see the coefficients for each variable and how (or if) they contribute to the overall sales price of a product.  We will then use the columns we find to be most impactful to the sales price as columns used in our prediction model as well.  

We chose columns based on how their addition to the linear regression model impacted R^2 and adjusted R^2.  If it did not increase, or even lowered either of those two statistics, then we did not keep it as a useful predictor column.  This is because R^2 is the coefficient of determination, is a statistical measure in regression analysis that shows the proportion of the variance in the dependent variable that is predictable from the independent variables. It indicates how well the model fits the data, with values from 0 to 1, where 1 means a perfect fit and 0 means the model explains none of the variability.  Adjusted R^2 penalizes the addition of irrelevant variables, and so it is even more sensitive than R^2 to non-useful contributors.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
import statsmodels.api as sm

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data/02-processed/merged.csv')
### renaming columns to remove spaces
df = df.rename(columns = {"Listing Price": "Listing_Price", "Sale Price":"Sale_Price", "Product Type":"Product_Type"})

### printing a statement to add context to why we are doing this (for hedonic analysis)
### we will only add columns which increase the R^2 and adjusted R^2, so as to not 
### unnecesarily complicate the model
print('The purpose of this analysis is to discover the weights/coefficients of each variable as \ndeterminants of sale price prior to using a machine learning model to predict for it.  \nThis aligns with our goal of a hedonic analysis of (sales) price, since it helps us determine \nindividual factors that impact sales price and gives us details that would be lost \nin a machine learning model.')

### OLS regression with just listing price to predict sales price
print('\n\nOrdinary Least Squares Regression - Using listing price to predict sales price, as a baseline model')
regression_one = smf.ols('Sale_Price ~ Listing_Price', data=df).fit()
print(regression_one.summary())

### OLS regression with listing price and product type to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price and product type to predict sales price\n')
regression_two = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type)', data=df).fit()
print(regression_two.summary())

### OLS regression with listing price, product type, and gender to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, and target gender to predict sales price')
regression_three = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender)', data=df).fit()
print(regression_three.summary())

### OLS regression with listing price, product type, gender, and country to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, target gender, and country \nto predict sales price')
regression_four = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender) + C(Country)', data=df).fit()
print(regression_four.summary())

### OLS regression with listing price, product type, gender, country, and rating to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, target gender, country, \nand rating to predict sales price')
regression_five_a = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender) + C(Country) + Rating', data=df).fit()
print(regression_five_a.summary())
### ratings did not increase R^2 or adjusted R^2, so we did not keep it


### OLS regression with listing price, product type, gender, country, and reviews to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, target gender, country, \nand Reviews to predict sales price')
regression_five_b = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender) + C(Country) + Reviews', data=df).fit()
print(regression_five_b.summary())
### reviews did not increase R^2 or adjusted R^2, so we did not keep it


### OLS regression with listing price, product type, gender, country, and popularity to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, target gender, country, \nand Popularity to predict sales price')
Cregression_five_c = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender) + C(Country) + C(Popularity)', data=df).fit()
print(regression_five_c.summary())
### popularity did not increase R^2 or adjusted R^2, so we did not keep it


### OLS regression with listing price, product type, gender, country, and color to predict sales price
print('\n\nOrdinary Least Squares Regression - using listing price, product type, target gender, country, \nand Color to predict sales price')
regression_five_d = smf.ols('Sale_Price ~ Listing_Price + C(Product_Type) + C(Gender) + C(Country) + C(Color)', data=df).fit()
print(regression_five_d.summary())
### color did not increase R^2 or adjusted R^2, so we did not keep it


print("\n\nWe see from this regression model that R^2 and the adjusted R^2 peak at 4 columns used \nfor predicting sales price (price point of purchase), and that adding more than \nthe initial 4 (Listing price, product type, gender, and country) actually reduced the \nadjusted R^2, meaning it was a worse fit with more than those 4 predictors.   \n\nSo, since the Ratings, Popularity, Color, and Review columns are not helpful for predictions, \nas seen by this preliminary regression, we will not be using these columns in our prediction models.\n\n")
```

### ANALYSIS 2A: PREDICTING SALES PRICE FROM PRODUCT ATTRIBUTES USING LINEAR REGRESSION MODELS

Here, we use linear regression models to predict sales price.  This includes a linear regression  based on the columns we have identified as having been most useful in our hedonic analysis, as well as a ridge regression model (a type of linear regression model) combined with grid search to see if it yields a better model.  

In the following code, we will just be predicting sales price with our basic linear regression model.  

```python

### copying the dataframe, just to be safe and not alter anything we don't want to
df_ml = df_model.copy()

### One-Hot Encoding the categorical data, using built-in pandas method
### for an easier time displaying the data later (used instead of sklearn's one-hot encoder)
df_encoded = pd.get_dummies(df_ml, columns=['Gender', 'Product Type', 'Country','Color', 'Popularity'], 
                            drop_first=True, dtype=int)

### Selecting columns for prediction, as determined by previous hedonic analysis
hedonic_feature_cols = [col for col in df_encoded.columns 
                if col.startswith('Gender_') or 
                   col.startswith('Product Type_') or 
                   col.startswith('Country_')]
hedonic_feature_cols.append('Listing Price')

### Splitting the data into variables and targets (what we want to predict)
X_hedonic = df_encoded[hedonic_feature_cols]
y = df_encoded['Sale Price']

### Splitting data into train and test sets, with a 70/30 training/testing split
X_train_hedonic, X_test_hedonic, y_train, y_test = train_test_split(X_hedonic, y, test_size=0.3, random_state=42)


#### Trying Model 1: Simple Linear Regression using columns we determined as significant 
#### (Hedonic Analysis)
print("\n\nPREDICTION MODEL 1: Linear Regression \n")
print("-"*80)
lr = LinearRegression()
lr.fit(X_train_hedonic, y_train)
y_pred_lr = lr.predict(X_test_hedonic)

r2_lr = r2_score(y_test, y_pred_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))

print(f"\nTest Set Performance:")
print(f"  R² Score: {r2_lr}")
print(f"  MAE: ${mae_lr}")
print(f"  RMSE: ${rmse_lr}")

### Showing the top 5 most important features, determined by this model
feature_importance = pd.DataFrame({
    'Feature': hedonic_feature_cols,
    'Coefficient': lr.coef_
})

feature_importance['Abs_Coefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values('Abs_Coefficient', ascending=False)
print('\n')
print("-"*80)
print("\nLinear Regression Model: Top 5 Most Important Features:")
print(feature_importance[['Feature', 'Coefficient']].head(5).to_string(index=False))


```

### ANALYSIS 2B: PREDICTING SALES PRICE FROM PRODUCT ATTRIBUTES USING LINEAR REGRESSION MODELS 

Here, we are still using linear regression models to predict sales price.  This includes a linear regression  based on the columns we have identified as having been most useful in our hedonic analysis, as well as a ridge regression model (a type of linear regression model) combined with grid search to see if it yields a better model.  

In the following code, we will be predicting sales price using our ridge regression model.  

```python
### printing out definition of ridge regression so it is clear what we are doing in-context
print("\n\nRidge regression is a regularized version of linear regression, meaning it adds a penalty term \nto normal linear regression, which can help make a more stable, generalizable model by penalizing \nmodel complexity.")
print("We are trying Ridge Regression with Grid Search, to see if it's an improvement on the \nLinear Regression model with our selected columns from the hedonic analysis.\n\n")

### Trying Model 2: Ridge Regression with Grid Search, to see if it's an improvement on previous model 
### (still using our selected columns from the hedonic analysis)
print("-"*60)
print("\nPREDICTION MODEL 2: Ridge Regression with Grid Search:\n")

#### Ridge regression is a regularized version of linear regression, meaning it
#### adds a penalty term to normal linear regression, which can help make a more 
#### stable, generalizable model by penalizing model complexity. 


### Defining a parameter grid for the grid search
param_grid = {
    'alpha': [0.1, 1.0, 10.0, 15]
#### alpha controls the strength of regularization, so can help prevent overfitting, depending on 
#### it's strength (so we are testing a range of alpha values)
}

ridge = Ridge()
grid_search1 = GridSearchCV(
    ridge, 
    param_grid, 
    cv=5, 
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search1.fit(X_train_hedonic, y_train)

print(f"\nBest alpha: {grid_search1.best_params_['alpha']}")
print(f"Best CV score (neg MSE): {grid_search1.best_score_}")

### Evaluating model on test set
y_pred_ridge = grid_search1.predict(X_test_hedonic)
r2_ridge = r2_score(y_test, y_pred_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))

print(f"\nTest Set Performance:")
print(f"  R^2 Score: {r2_ridge}")
print(f"  MAE: ${mae_ridge}")
print(f"  RMSE: ${rmse_ridge}")

### Showing the top 5 most important features, determined by this model
feature_importance_rr = pd.DataFrame({
    'Feature': hedonic_feature_cols,
    'Coefficient': grid_search1.best_estimator_.coef_

})

feature_importance_rr['Abs_Coefficient'] = feature_importance_rr['Coefficient'].abs()
feature_importance_rr = feature_importance_rr.sort_values('Abs_Coefficient', ascending=False)
print('\n')
print("-"*60)
print("\nRidge Regression Model: Top 5 Most Important Features:")
print(feature_importance_rr[['Feature', 'Coefficient']].head(5).to_string(index=False))



### Comparing Ridge Regression model to Linear Regression model
print('\n')
print("-"*60)
print(f"\nImprovement (or not) over Linear Regression:")
print(f"  R^2 change: {r2_ridge - r2_lr}")
print(f"  MAE change: ${mae_ridge - mae_lr}")
if r2_ridge - r2_lr < 0:
    print("  Conclusion: The Ridge Regression Model was not an improvement over our Linear Regression Model for \npredicting sales price\n")
elif r2_ridge - r2_lr > 0:
    print("  Conclusion: The Ridge Regression Model was an improvement over our Linear Regression Model for \npredicting sales price\n")


print("-"*60)

### Visualizing our predictions
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

#### looking at our Linear Regression predictions
axes[0].scatter(y_test, y_pred_lr, alpha=0.5, s=20)
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
            'r--', lw=2, label='Perfect Prediction')
axes[0].set_xlabel('Actual Sale Price ($)', fontsize=12)
axes[0].set_ylabel('Predicted Sale Price ($)', fontsize=12)
axes[0].set_title(f'Linear Regression\n R^2 = {r2_lr}', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

#### looking at our Ridge Regression Predictions vs our Linear Regression predictions
axes[1].scatter(y_test, y_pred_ridge, alpha=0.5, s=20, color='green')
axes[1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
            'r--', lw=2, label='Perfect Prediction')
axes[1].set_xlabel('Actual Sale Price ($)', fontsize=12)
axes[1].set_ylabel('Predicted Sale Price ($)', fontsize=12)
axes[1].set_title(f'Ridge Regression (alpha ={grid_search.best_params_["alpha"]})\nR^2 = {r2_ridge}', 
                 fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.savefig('results/model_predictions_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

```

### ANALYSIS 3: PREDICTING SALES PRICE FROM PRODUCT ATTRIBUTES USING NONLINEAR REGRESSION

Random forest regression is not linear, as it fits a number of decision tree regressors on many sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting, in the hopes of being a good way of predicting. 

We are pursuing this (a non-linear regression model) to see if our hedonic analysis can be improved, and see what regression model could best assess factors determining sales price.  This is because we have essentially just been using linear regression for predictions, which may not be able to take into account more complex factors affecting sales price. ## Discussion and Conclusion

### Summary of Results
This study investigated the factors influencing athleticwear pricing through a hedonic analysis of Adidas products in the U.S. market. Our analysis revealed several significant findings that both challenged and extended existing research. First, we found that listing price was by far the strongest predictor of sale price (importance score: 0.986 in our Random Forest model), serving as the baseline reference point from which other product attributes add or subtract value. Beyond listing price, product type and gender-target emerged as the most significant pricing determinants, with footwear being tied to substantially higher prices than clothing or accessories, and gender-based pricing disparities modestly present across categories. Specifically, women's products showed 1.94% higher listing prices and 2.07% higher sale prices compared to men's products on average, while unisex products commanded the highest prices overall (18.4% higher sale prices than men's products, and 16% higher than women's products). However, it should be noted that the sample of unisex products was much smaller than either men's products or women's products in our analysis, and so it might not be an accurate representation of unisex pricing in general, especially if it consisted of mainly footwear, the most costly product type on average (as seen in the plots). It should also be noted that our dataset was relatively small overall, so these conclusions are not a good representation of all gender pricing distribtuons. However, even with this limited data, we saw that country of origin showed notable influence, with products from "Various" countries priced approximately 5-6 dollars lower than U.S.-made products, controlling for other factors. Our data lacks the detail we would need to distinguish an actual cause or stronger evidence, but the consistent negative coefficient (-5 to -6 dollars) suggests that country of origin is a real, reproducible pattern worthy of future investigation.  So, overall, product type, gender, and country were our most significant predictors, with listing price acting as our baseline price.  We also found that variables we expected to matter, such as color, customer ratings, review counts, and our derived popularity metric, showed minimal predictive power in our models, so our hypothesis was incorrect.  


### Extension of Prior Research 
Our findings both align with and diverge from the existing research in important ways. The hedonic pricing methodology we used follows the established tradition used by economists and government agencies to decompose product prices into constituent attributes, as described in our background research on the Consumer Price Index. However, our results didn't completely align with some claims made in prior work. While Lindahl's (2018) study of outdoor apparel found material type to be a statistically significant pricing driver, we were unable to evaluate this hypothesis due to severe data limitations, as we did not have material composition data in our datasets. This was a critical missing piece, as the background research strongly suggested that material (synthetic vs. natural fibers, organic cotton, technical fabrics) were fundamental factors in both production costs and consumer willingness to pay in the athleticwear market. On the other hand, our findings on gender-based pricing were at least able to be compared to the present information to the "Pink Tax" debate discussed in our background section. The U.S. Congress Joint Economic Committee (2016) and NYC Department of Consumer Affairs (2015) documented systematic price premiums for women's products, while Moshary et al. (2023) argued that many apparent disparities reflect product differentiation rather than discrimination, but our results weren't significant enough to be conclusive.  Women's athleticwear from Adidas shows modest listing and sales price premiums (1.94% and 2.07%, respectively), but these increases are too small to determine whether these are results of the "pink tax", or just due to random chance. However, the fact that gender remained a statistically significant predictor even after accounting for product type, country, and listing price suggests that gender-targeted marketing does influence final pricing beyond mere product differences. We also though that the relative unimportance of popularity metrics (ratings, reviews) in our models was unexpected, given the background research emphasizing their influence. Northwestern's research found that products with five reviews had 270% greater purchase likelihood, and Sung et al. (2023) found that consumers considered star ratings more important than price. The disconnect between these factors' importance for purchase decisions versus price determination suggests that Adidas might use  standardized pricing strategies that don't account for product popularity, or that our popularity measures (deviation from mean reviews for product type) were not good enough to capture the true signals that drive pricing decisions.  


### Contradictions, Complexities, and Points of Confusion
Several apparent contradictions emerged that warrant discussion. Our exploratory data analysis revealed that when analyzing products by gender and product type simultaneously, women's products  showed higher or equal prices to men's across all three apparel categories (clothing, footwear, accessories). However, when we aggregated across all categories, the differences became smaller. This suggests Simpson's Paradox may be at play: the relationship between gender and price appears to change when we control for product type. This could be due to the distribution of products, as if women's products are disproportionately concentrated in lower-priced categories (like more clothing items than footwear), the aggregate averages can be misleading. This is very much possible, as athleticwear is perhaps the one category of clothing that targets more men than women.  Another interesting finding was the strong negative coefficient for "Various" countries in our regression models. Initially, we hypothesized that country of origin would matter due to the COO effect described by Diamantopoulos et al. (2012), where favorable country associations command price premiums. Our analysis showed that products from "Various" countries were consistently cheaper. This could reflect several realities, such as "Various" indicating multi-country manufacturing typical of lower-cost production chains (like production being outsourced to Vietnam or China), consumers discounting products without clear origins, or this variable being confounded with other variables, like product type (some products can be made with parts from multiple places). Our data limitations prevent us from definitively resolving this question. 

### Limitations and Biases 
Our study faces several significant limitations that prevent us from being able to generalize our results and interpretations. Most critically, our analysis is limited exclusively to Adidas products—we cannot project our findings onto Nike, Under Armour, Lululemon, or the broader athleticwear industry, each of which likely employs distinct pricing strategies. This would also help with our issue of not having enough data, as our dataset isn't very large overall, and is unbalanced when it comes to gender distributions (many more men's products than women's products, and there are relatively few unisex products in comparison to those both). The datasets we used also introduce several forms of bias. Selection bias, as online-only products may differ from in-store inventory, temporal bias, as cross-sectional data from a single time period cannot capture seasonal pricing dynamics or trend evolution, and brand bias, as Adidas occupies a specific market position between mass-market and premium brands. Also, the lack of material composition data is a huge limitation, essentially preventing us from testing our primary hypothesis. Given the background research's emphasis on material as a fundamental pricing driver—with production costs, sustainability premiums, and performance characteristics all varying dramatically by fabric type, our inability to include this variable could be an explanation behind the modest R^2 scores (~0.785) in our best predictive model. THis R^2 means we're only explaining only about 78.5% of price variance, which suggests that unmeasured variables (which almost certainly includes material) account for the remaining approximate 21.5%. Additionally, our "popularity" metric (deviation from mean reviews by product type) was not sophisticated enough to capture the complex dynamics described in the e-commerce background research. If we had more detailed data, such as individual SKU-level sales volumes, time-series pricing data to track how prices evolve, material composition, detailed product specifications (cushioning technology, waterproofing, etc.), and competitor pricing for similar items, we would probably be able to conduct a better, more conclusive study. Longitudinal data would be particularly interesting if we wanted to investigate whether gender-based pricing disparities are changing over time (widening, narrowing, or stable). This would be particularly useful to see if the "pink tax" exists, or had ever existed, and if it had, then using that data to seee how or if it's changed due to awareness of pricing inequity would be interesting. 


### Future Directions 
If this work were to continue, several extensions would be valuable. First, expanding to multi-brand analysis would reveal whether the patterns we observed are Adidas-specific or industry-wide, with particular interest in comparing Adidas to direct competitors like Nike and niche players like Lululemon who position differently on gender-inclusivity. Even expanding the data to an even broader market, such as all major apparel retailers, could make this a study much more generalizable beyond athleticwear, and might reveal patterns otherwise unseen.  Second, acquiring material composition data, even if through manual coding of product descriptions, would allow us to properly test our original hypothesis and likely improve model performance substantially. Third, actual tests with human subjects could help test for pink tax, if we ran a study seeing whether consumers actually pay more for products labeled "women's" versus identical items labeled "unisex".  Finally,  tracking pricing changes over time in response to public discussion of the Pink Tax could reveal whether social pressure or other causes drive pricing equity over time. 

### Overall Impact 
Despite these limitations, our study makes a valuable contribution in that it demonstrates how product category and gender-target are more immediately relevant to athleticwear pricing than some factors emphasized in prior research (like popularity metrics).  The shortcomings of our study also highlight the need for more data (such as material data) in a comprehensive hedonic analysis of apparel pricing. Regardless, our findings suggest that both product type (footwear vs. clothing) and gender-based marketing both impact the athleticwear pricing landscape, with implications for both consumer awareness and potential policy interventions around pricing equity.
